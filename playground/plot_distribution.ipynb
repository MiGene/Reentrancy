```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import glob
import os
```

```python
TRUE_COL = 'is_sus'
PRED_COL = 'predicted'
EXCEL_FOLDER = './test_attack_feature_sim/'
CSV_FOLDER = './test_attack_result/'
```

```python
def plot_distribution(df,features,filename):
    i = 1
    plt.figure(figsize=(20,4))  # Adjust width and height for proper spacing
    for feature in features:
        tp_df = df[(df[TRUE_COL]==1) & (df[PRED_COL]==1)][feature]
        fp_df = df[(df[TRUE_COL] == 0) & (df[PRED_COL] == 1)][feature]
        tn_df = df[(df[TRUE_COL] == 0) & (df[PRED_COL] == 0)][feature]
        # Assuming you have a DataFrame 'df' with columns 'A' and 'B'
    
        plt.subplot(1, 5, i)  # (1 row, 5 columns, index i)
        # Plotting the distribution of column A
        # sns.kdeplot(tp_df, shade=True, label="True positive")
        # # Plotting the distribution of column B
        # sns.kdeplot(fp_df, shade=True, label="False positive")
        sns.kdeplot(tn_df, shade=True, label="True negative")
        for tp in tp_df.to_list():
            plt.axvline(x=tp, color='red', linestyle='-', label=f'True positive at {tp}')

        for fp in fp_df.to_list():
            plt.axvline(x=fp, color='blue', linestyle='--', label=f'False positive at {fp}')


        # Adding labels and title
        plt.title(f'{feature}, {filename}')
        plt.xlabel('Values')
        plt.ylabel('Density')
        # plt.legend()

        i += 1
    plt.tight_layout()
        # Show the plot
    plt.show()

def get_top_5_distances(path,unwanted_features):
    df = pd.read_excel(path)
    ranked_features = df['Feature'].to_list()
    ranked_features = [f for f in ranked_features if f not in unwanted_features]
    print(ranked_features)
    return ranked_features[0:5]
```

```python
unwanted_features = [  # for n=11
        "contract_lifetime_block",
        "contract_lifetime_days",
        "sender_tx_count_call_contract",
        "sender_lifetime_days",
        "sender_lifetime_block",
        "contract_main_active_days",
        "value",
        "contract_block_ratio",
        "tx_sender_call_contract",
        "predicted",
        "is_sus",
    ]


top_5_importance = pd.read_csv('./feature_importance.csv')['Feature'].to_list()[0:5]

print(glob.glob(os.path.join(EXCEL_FOLDER, "*.xlsx")))
for file_path in glob.glob(os.path.join(EXCEL_FOLDER, "*.xlsx")):
    # print(1)
    filename = os.path.basename(file_path)
    if 'feature_order' not in filename: continue
    if '~$' in filename: continue
    top_5_distances = get_top_5_distances(file_path,unwanted_features)
    csv_filename = filename.replace('feature_order_','').replace('.xlsx','.csv')
    csv_path = f'{CSV_FOLDER}{csv_filename}'
    df = pd.read_csv(csv_path)
    plot_distribution(df,top_5_importance,csv_filename)

    
```
